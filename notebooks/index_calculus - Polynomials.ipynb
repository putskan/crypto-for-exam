```python
# this one is only pretty partial. will probably get half the points

####### config start #######

# example taken from the last lab, Q6

mod = 2
P.<y> = PolynomialRing(GF(mod))
F29.<x> = GF(mod).extension(y^9 + y^4 + 1)
highest_degree = 9
g = x
g_a = x^4 + x

# taking some factor base:
factor_base = [
    x + 1,
    x^4 + x + 1,
    x^2 + x + 1
]

####### config end #######


P.<y> = PolynomialRing(GF(mod))
F29.<x> = GF(mod).extension(y^9 + y^4 + 1)

PZ.<X> = PolynomialRing(ZZ)
factor_base = [PZ(f) for f in factor_base]


useful_factorizations = []
for a in range(1, 100):
    factorization = PZ(g^a).factor()
    print(f'g^{a} =', factorization)
    is_useful_factorization = True
    
    for f in factorization:
        curr_poly = f[0].change_ring(Zmod(mod))
        if curr_poly not in factor_base:
            is_useful_factorization = False
            print(f[0], 'is not in factor base. not useful')
            break
            
    if is_useful_factorization:
        useful_factorizations.append((a, factorization))
    
    if len(useful_factorizations) == len(factor_base):
        print(f'\n\nfound {len(factor_base)} useful equations')
        break
        
    print()

print('########################################')
print('\nEquations found:')
for a, factorization in useful_factorizations:
    print(f'g^{a} = {factorization}')
    
assert len(useful_factorizations) == len(factor_base), f'found only {len(useful_factorizations)}/{len(factor_base)} factorizations'


print('\n')
print('Taking log we get:')
for a, factorization in useful_factorizations:
    eq_str = fr'{a} = '
    for poly, power in factorization:
        eq_str += fr'{power} \cdot log_{{{g}}}({{{poly}}}) + '
    eq_str = eq_str[:-2]
    eq_str += fr' \;\;\; (mod \; {mod}^{highest_degree} - 1)'
    display(Math(eq_str))


    

    
print('\n\n*** Now we need to solve manually, and after that, factorize the original g^a ***\n\n')

```
